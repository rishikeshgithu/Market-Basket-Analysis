```python
import pandas as pd

# Function to load data
def load_data(file_path):
    data = pd.read_excel(file_path)
    return data

# Function to count frequencies of items bought with antecedent
def count_items_bought_with_antecedent(data, item_type, selected_entity):
    items_bought_with = {
        'sku_name': {},
        'brand_name': {},
        'Category': {}
    }

    # Filter transactions where antecedent (selected_entity) appears
    selected_item_data = data[data[item_type] == selected_entity]
    selected_item_receipt_ids = selected_item_data['receipt_id'].unique()

    # Iterate through each transaction where antecedent appears
    for receipt_id in selected_item_receipt_ids:
        receipt_data = data[data['receipt_id'] == receipt_id]
        receipt_data = receipt_data[receipt_data[item_type] != selected_entity]  # Exclude the antecedent itself
        for entity in items_bought_with.keys():
            entities = receipt_data[entity].unique()
            for item in entities:
                if item in items_bought_with[entity]:
                    items_bought_with[entity][item] += 1
                else:
                    items_bought_with[entity][item] = 1

    return items_bought_with

# Load data
data = load_data("Sample_Transaction_Data.xlsx")

# Set parameters for analysis
item_type = 'sku_name'  # Example: 'sku_name'
selected_entity = 'Prod_1'   # Example: 'Prod_1'

# Step 1: Count the frequency of the antecedent
antecedent_frequency = data[data[item_type] == selected_entity].shape[0]
print(f"Frequency of antecedent '{selected_entity}' ({item_type}): {antecedent_frequency}")

# Step 6: Count the frequency of all sku_name, brand_name, and Category bought with the antecedent
items_bought_with = count_items_bought_with_antecedent(data, item_type, selected_entity)

# Print results
print(f"\nFrequency of {item_type} bought with '{selected_entity}':")
for sku_name, count in items_bought_with['sku_name'].items():
    print(f"{sku_name}: {count}")

print(f"\nFrequency of brand_name bought with '{selected_entity}':")
for brand_name, count in items_bought_with['brand_name'].items():
    print(f"{brand_name}: {count}")

print(f"\nFrequency of Category bought with '{selected_entity}':")
for category, count in items_bought_with['Category'].items():
    print(f"{category}: {count}")
```
